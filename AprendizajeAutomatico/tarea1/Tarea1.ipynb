Creamos una nueva variable categórica para indicar el tipo de clima de acuerdo a temperatura y precipitación

```python
# Aplicar la clasificación del clima correctamente
def clasificar_clima(temp, precip):
    if pd.isnull(temp) or pd.isnull(precip):
        return "Desconocido"
    if temp < 15 and precip > 20:
        return "Frío y Lluvioso"
    elif temp < 15 and precip <= 20:
        return "Frío y Seco"
    elif 15 <= temp <= 25 and precip > 20:
        return "Templado y Lluvioso"
    elif 15 <= temp <= 25 and precip <= 20:
        return "Templado y Seco"
    elif temp > 25 and precip > 20:
        return "Cálido y Lluvioso"
    else:
        return "Cálido y Seco"

# Crear nueva columna categórica
df_original_cleaned["Clasificacion_Clima"] = df_original_cleaned.apply(
    lambda row: clasificar_clima(row["Temperatura_Agua_C"], row["Precipitacion_mm"]), axis=1
)
```

```python
# Definir un umbral para eliminar filas con demasiados valores faltantes (ejemplo: más del 50% de las columnas con NaN)
threshold = len(df_original_cleaned.columns) * 0.5

# Filtrar filas donde el número de valores no nulos sea mayor al umbral
df_filtered = df_original_cleaned.dropna(thresh=threshold).reset_index(drop=True)
```

```python
# Excluir la columna de fecha antes de calcular la media
df_imputada_con_media = df_filtered.drop(columns=["FechaHora"]).copy()

# Seleccionar solo columnas numéricas para imputación
numeric_cols = df_filtered.select_dtypes(include=['number']).columns

# Reemplazar valores nulos con la media en las columnas numéricas
df_imputada_con_media[numeric_cols] = df_imputada_con_media[numeric_cols].apply(lambda x: x.fillna(x.mean()))
```

```python
# Excluir la columna de fecha antes de calcular la media y la mediana
df_imputada_con_mediana = df_filtered.drop(columns=["FechaHora"]).copy()

# Seleccionar solo columnas numéricas para imputación
numeric_cols = df_filtered.select_dtypes(include=['number']).columns

# Reemplazar valores nulos con la mediana en las columnas numéricas
df_imputada_con_mediana[numeric_cols] = df_imputada_con_mediana[numeric_cols].apply(lambda x: x.fillna(x.median()))
```

**Objetivo:** Identificar y manejar valores atípicos en el dataset.

Para detectar outliers, es fundamental usar un dataset que no haya sido normalizado o estandarizado, ya que MinMaxScaler y StandardScaler alteran la distribución de los datos, lo que podría dificultar la identificación de valores extremos.

**df_original_cleaned**:: No se recomienda ya que contiene valores nulos y errores en los datos.

**df_imputada_con_mediana**:: Parcialmente útil	algunos valores extremos pueden haberse suavizado.

**df_imputada_con_media**:: Parcialmente útil	sensible a outliers, podría haber valores distorsionados.

**df_mediana_minmax**:: No recomendado,	todos los valores están escalados entre 0 y 1, lo que dificulta detectar outliers.

**df_interpolacion_standard**:: No recomendado,	La media es 0 y la varianza es 1, lo que normaliza outliers y los oculta.

**df_imputada_con_interpolacion**:: Mejor opción	ya que conserva la estructura y la distribución original de los datos, no ha sido escalada, lo que evita distorsiones en la detección de valores extremos.
No tiene valores nulos, lo que facilita aplicar métodos como IQR (rango intercuartílico), Z-score o DBSCAN.

**Como alternativa**, en nuestro caso, el dataset tiene datos secuenciales de clima (o sensores), por lo que también podría usarse df_imputada_con_mediana si los outliers son más relevantes a nivel de variación de datos.


Dado que nuestro dataset contiene variables meteorológicas y ambientales como temperatura, precipitación, presión, humedad, etc., la elección del mejor método para detectar outliers depende de la naturaleza de los datos.

**La mejor opción en nuestro caso es el método IQR** (Rango Intercuartílico) porque:

Nuestros datos no necesariamente siguen una distribución normal.
Tenemos variables continuas como temperatura, precipitación y presión, donde los valores atípicos pueden ser errores o eventos extremos.
IQR no es sensible a valores extremos individuales y no requiere suposiciones sobre la distribución de los datos.
Es fácil de interpretar y aplicar sin requerir ajustes complejos como DBSCAN.

**¿Cuándo considerar Z-score o DBSCAN?**

**Z-score::** Si verificamos que algunas variables tienen una distribución normal o queremos identificar valores extremos en un contexto más estadístico.

**DBSCAN::** Si los datos tienen patrones de agrupación claros (como eventos climáticos extremos en ciertas regiones o periodos de tiempo específicos).

**Aplicar el método IQR para detectar outliers estadísticamente**

**Objetivo:** Detectar valores fuera del rango esperado.

```python
# Select only numeric columns before calculating quantiles
numeric_columns = df_imputada_con_interpolacion.select_dtypes(include=['number']).columns
Q1 = df_imputada_con_interpolacion[numeric_columns].quantile(0.25)
Q3 = df_imputada_con_interpolacion[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Apply outlier detection only to numeric columns
outliers = (df_imputada_con_interpolacion[numeric_columns] < lower_bound) | (df_imputada_con_interpolacion[numeric_columns] > upper_bound)

# Filter the original DataFrame based on outliers in numeric columns
df_outliers = df_imputada_con_interpolacion[outliers.any(axis=1)]

df_outliers.describe()
```

Extracción los Outliers para eventos extremos

```python
# Crear una copia del dataset interpolado
df_extreme_events = df_imputada_con_interpolacion.copy()

# Seleccionar solo las columnas numéricas
numeric_columns = df_extreme_events.select_dtypes(include=['float64', 'int64']).columns

# Calcular IQR
Q1 = df_extreme_events[numeric_columns].quantile(0.25)  # Primer cuartil
Q3 = df_extreme_events[numeric_columns].quantile(0.75)  # Tercer cuartil
IQR = Q3 - Q1  # Rango intercuartílico

# Definir umbrales más estrictos para detectar solo eventos extremos (outliers fuertes)
lower_extreme_bound = Q1 - 3 * IQR  # Se usa 3 * IQR en lugar de 1.5 para detectar eventos realmente extremos
upper_extreme_bound = Q3 + 3 * IQR

# Crear una máscara para identificar solo los eventos climáticos extremos
extreme_mask = (df_extreme_events[numeric_columns] < lower_extreme_bound) | (df_extreme_events[numeric_columns] > upper_extreme_bound)

# Filtrar el dataset con solo los eventos extremos detectados
df_extreme_outliers = df_extreme_events[extreme_mask.any(axis=1)]

df_extreme_outliers.head()

```

# **tratamientode los outliers**

**Si los valores son errores de medición**:: o datos imposibles, (ejemplo: temperatura negativa en una zona tropical), esos deben ser eliminados o corregidos.

**Si los valores son eventos climáticos extremos**:: se mantienen en un dataset separado para análisis de anomalías.

**Si hay dudas sobre un valor:** se pueden reemplazar con mediana o valores interpolados.

En nuestro caso específico,

**para la mayoría de las variables ambientales**
(ejemplo: temperatura, presión, humedad), usar IQR para detectar valores fuera de rango.

**Para eventos extremos**
(ejemplo: lluvias intensas, olas de calor), se deben conservar los outliers y analizarlos por separado.


**tratamiento de Outliers no-extremos**

Para la identificación de Outliers, **IQR** es más recomendado porque no depende de la distribución de los datos y funciona bien con valores sesgados.
**Z-score** no es la mejor opción porque nuestro dataset no tiene distribución normal y podríamos eliminar eventos importantes.

Si se utiliza un **IQR 1.5**, los valores identificados pueden ser considerados atípicos, pero no necesariamente errores, estos valores se deben reemplazar con la mediana para evitar distorsión en el análisis sin perder la estructura de los datos, mantiene la tendencia general del dataset y es menos sensible a outliers que la media.

Si se utiliza un **IQR 3**, los valores identificados son valores extremos que deben manejarse por separado ya que podrían representar alteraciones extremas del clima y estadísticamente pueden ser de utilidad. Pero para ser utilizado como entrada en un modelo de ML, no tendrían mucha utilidad, afectarían el proceso de prediccion. Por lo que debn ser eliminados.

```python
# Crear copia del dataset con interpolación
df_final = df_imputada_con_interpolacion.copy()

# Definir condiciones de errores físicos
errores_condiciones = (
    (df_final["Temperatura_Agua_C"] < -30) | (df_final["Temperatura_Agua_C"] > 60) |
    (df_final["Temperatura_Interna_C"] < -10) | (df_final["Temperatura_Interna_C"] > 50) |
    (df_final["Humedad_Interna_Porc"] < 0) | (df_final["Humedad_Interna_Porc"] > 100) |
    (df_final["Presion_Sensor_cm"] < 0) |
    (df_final["Bateria_Volts"] < 10) | (df_final["Bateria_Volts"] > 15) |
    (df_final["Corriente_Panel_Solar"] < 0) |
    (df_final["Precipitacion_mm"] < 0) |
    (df_final["Precipitacion_24h_mm"] < 0) | (df_final["Precipitacion_24h_mm"] > 500) |
    (df_final["Nivel_Agua_m"] < 0)
)

# Eliminar los valores fuera del rango físico
df_final = df_final[~errores_condiciones]

# Eliminar valores extremos según IQR 3
if 'df_extreme_outliers' in locals():
    df_final = df_final[~df_final.index.isin(df_extreme_outliers.index)]

# Aplicar interpolación en los outliers moderados (IQR 1.5)
if 'df_outliers' in locals():
    # Get the common indices between df_final and df_outliers
    common_indices = df_final.index.intersection(df_outliers.index)

    # Interpolate only on the common indices and columns
    df_final.loc[common_indices, df_outliers.columns] = df_final.loc[common_indices, df_outliers.columns].interpolate(method='linear')

df_final.head()
```
